```python
temp = np.array([36.8, 37.1, 36.9, 37.5, 38.2, 36.7, 37.0])
heart = np.array([62, 70, 65, 74, 90, 60, 68])
```

```python
temp_cr = (temp - temp.mean()) / temp.std()
heart_cr = (heart - heart.mean()) / heart.std()
```

```python
import matplotlib.pyplot as plt

"""
quand la température est anormalement élevée (au sens statistique), la fréquence cardiaque tend aussi à être anormalement élevée.
Et inversement.

"""

x = np.arange(len(temp_cr))

# Tracé comparatif
plt.figure()
plt.scatter(x, temp_cr, label="Température (CR)")
plt.scatter(x, heart_cr, label="Fréquence cardiaque (CR)")
# plt.vlines(x, temp_cr, heart_cr)  # écart point à point
plt.axhline(0) # moyenne on la met pour plus de lisibilité sur la dispersion

plt.xlabel("Observation")
plt.ylabel("Valeur centrée-réduite")
plt.title("Comparaison après centrage-réduction")
plt.legend()
plt.show()
```

```python


# Données brutes
temp = np.array([36.8, 37.1, 36.9, 37.5, 38.2, 36.7, 37.0])
heart = np.array([62, 70, 65, 74, 90, 60, 68])

# Axe des observations
x = np.arange(len(temp))

# Tracé comparatif (valeurs non centrées-réduites)
plt.figure()
plt.scatter(x, temp, label="Température (°C)")
plt.scatter(x, heart, label="Fréquence cardiaque (bpm)")
# plt.vlines(x, temp, heart)  # écart point à point
plt.axhline(0)

plt.xlabel("Observation")
plt.ylabel("Valeur brute")
plt.title("Comparaison sur valeurs non centrées-réduites")
plt.legend()
plt.show()
```
