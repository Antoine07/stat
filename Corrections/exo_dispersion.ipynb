```python
A = np.array([
    45, 46, 47, 48, 49, 50, 50, 50, 51, 52, 53, 54, 55
])

B = np.array([
    45, 46, 47, 48, 49, 50, 50, 50, 51, 52, 53, 54, 55,
    80, 85, 90
])

C = np.array([
    40, 42, 44, 46, 48, 50, 52, 54, 56, 58, 60
]) 
```

1. Calculer la moyenne de A, B et C.
1. Calculer la médiane de A, B et C.
1. Comparer moyenne et médiane pour chaque série.
1. Calculer l’étendue de chaque série.
1. Calculer l’écart-type de A, B et C.
1. Classer les séries de la moins dispersée à la plus dispersée.
1. Tracer les histogrammes des trois séries.
1. Décrire la forme de chaque distribution
1. Tracer un diagramme en boîte comparatif pour A, B et C.
1. En quelques phrases, expliquer le lien entre : dispersion, asynmétrie, valeurs atypiques

```python
# 1. 2. 3. 4. 5.
mean_A = A.mean()
median_A = np.median(A)
std_A = A.std()
range_A = A.max() - A.min()

print(f"A -> moyenne: {mean_A :.2f} médiane: {median_A:.2f}, étendue: {range_A:.2f}, écart-type: {std_A:.2f}")
```

```python
# 1. 2. 3. 4. 5.
mean_B = B.mean()
median_B = np.median(B)
std_B = B.std()
range_B = B.max() - B.min()

print(f"A -> moyenne: {mean_B :.2f} médiane: {median_B:.2f}, étendue: {range_B:.2f}, écart-type: {std_B:.2f}")
```

```python
# 1. 2. 3. 4. 5.
mean_C = C.mean()
median_C = np.median(C)
std_C = C.std()
range_C = C.max() - C.min()

print(f"A -> moyenne: {mean_C :.2f} médiane: {median_C:.2f}, étendue: {range_C:.2f}, écart-type: {std_C:.2f}")
```

```python
# 6. Classer les séries de la moins dispersée à la plus dispersée.

dispersion = { "A" : std_A, "B" : std_B, "C" : std_C }
classement = sorted(dispersion.items(), key = lambda x: x[1] ) # ordre en fonction du tuple
```

```python
classement
```

```python
# Autre manière d'ordonner la dispersion, on obtiendra ici que les lettres 
classement = sorted(dispersion, key=lambda k: dispersion[k])
```

```python
classement
```

```python
# Tracer les histogrammes des trois séries.
plt.figure(figsize=(12,4))
plt.subplot(1,3,1) # ligne nombre de colonne position 
plt.hist(A, bins=6) # bins nombre de barres 
plt.title("A")

plt.subplot(1,3,2) # ligne nombre de colonne position 
plt.hist(B, bins=6) # bins nombre de barres 
plt.title("B")

plt.subplot(1,3,3) # ligne nombre de colonne position 
plt.hist(C, bins=6) # bins nombre de barres 
plt.title("C")

plt.show()
```

```python
# Forme
from scipy.stats import skew, kurtosis

print(f"A skew={skew(A)}, kurtosis={kurtosis(A)}")
print(f"C skew={skew(C)}, kurtosis={kurtosis(C)}")
print(f"B skew={skew(B)}, kurtosis={kurtosis(B)}")
```

```python
plt.figure()
plt.boxplot([A, B, C])

plt.title("Comparaison des groupes A, B et C (boxplot)")
plt.ylabel("Valeurs")

plt.show()
```
